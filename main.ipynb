# Analisis de Datos

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el dataset
data = pd.read_csv('data/Titanic - train.csv')
data.isna().sum()
```

```python
data.head()
```

```python
# El atributo 'Age' tiene valores faltantes. 
f, ax = plt.subplots(1, 2, figsize=(20, 10))

data['Survived'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Survived')
sns.countplot(data=data, x='Survived', ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```

```python
# Tres tipos de datos: 
# 1. categoricos
# 2. ordinales
# 3. continuas (i.g 0, 1, 2.36, 10000 etc)

# Datos Categoricos

# SEX
data.groupby(['Sex', 'Survived'])['Survived'].count()
f, ax = plt.subplots(1, 2, figsize=(20, 10))
data[['Sex', 'Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0])
sns.countplot(data=data, x='Sex', hue='Survived', ax=ax[1])
plt.show()

# El atributo 'Sex' tiene un atrivutir en la superviviencia 
```

```python
data.groupby(['Pclass', 'Survived'])['Survived'].count()

f, ax = plt.subplots(1, 3, figsize=(15, 5))
data['Pclass'].value_counts().plot.bar(ax=ax[0])
data[['Pclass', 'Survived']].groupby(['Pclass']).mean().plot.bar(ax=ax[1])
sns.countplot(data=data, x='Pclass', hue='Survived', ax=ax[2])
plt.show()
```

```python
# Redundacia de datos -> Independencia de las variables

# DATOS ORDINALES

sns.catplot(data=data, x='Pclass', y='Survived', hue='Sex', kind='point')
```

```python
# DATOS CONTINUOS

# AGE

print(f'{data['Age'].min()} años')
print(f'{data['Age'].mean()} años')
print(f'{data['Age'].max()} años')
```

```python
f, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.violinplot(data=data, x='Pclass', y='Age', hue='Survived', split=True, ax=ax[0])
sns.violinplot(data=data, x='Sex', y='Age', hue='Survived', split=True, ax=ax[1])
```

```python
fig, ax = plt.subplots(3, 3, figsize=(18, 18))

sns.countplot(data=data, x='Survived', ax=ax[0, 0])
ax[0, 0].set_title('Survived')

sns.countplot(data=data, x='Pclass', ax=ax[0, 1])
ax[0, 1].set_title('Pclass')

sns.countplot(data=data, x='Sex', ax=ax[0, 2])
ax[0, 2].set_title('Sex')

sns.histplot(data=data, x='Age', kde=True, ax=ax[1, 0])
ax[1, 0].set_title('Age')

sns.countplot(data=data, x='SibSp', ax=ax[1, 1])
ax[1, 1].set_title('SibSp')

sns.countplot(data=data, x='Parch', ax=ax[1, 2])
ax[1, 2].set_title('Parch')

sns.histplot(data=data, x='Fare', kde=True, ax=ax[2, 0])
ax[2, 0].set_title('Fare')

sns.countplot(data=data, x='Embarked', ax=ax[2, 1])
ax[2, 1].set_title('Embarked')

sns.boxplot(data=data, x='Pclass', y='Fare', ax=ax[2, 2])
ax[2, 2].set_title('Fare vs Pclass')

plt.tight_layout()
plt.show()
```
